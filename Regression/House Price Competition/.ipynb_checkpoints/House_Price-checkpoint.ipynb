```python
for f in ["LotFrontage","GarageYrBlt","MasVnrArea"]:
    data[f + "_nan"] = np.where(data[f].isnull(),1,0)
    data[f].fillna(data[f].median(),inplace=True)
```

```python
data
```

```python
for f in year[:-1]:
    data[f] = data["YrSold"] - data[f]
```

```python
l = []
for f in con:
    if 0 in data[f].unique():
        pass
    else:
        l.append(f)
```

```python
l
```

```python
for f in l:
    data[f] = np.log(data[f])
```

```python
for f in cats:
    temp = data.groupby(f).count()/data.shape[0]
    temp = temp[temp>0.01].index
    np.where(data[f].isin(temp),data[f],"Rare_var")
```

```python
data
```

```python
data.isnull().sum().sum()
```

```python
for f in cats:
    labels = data.groupby(f)["SalePrice"].mean().sort_values().index
    labels = {k:i for i,k in enumerate(labels)}
    data[f] = data[f].map(labels)
```

```python
data
```

```python
sc_features = [f for f in data.columns if f not in ["Id","SalePrice"]]
```

```python
sc = MinMaxScaler(feature_range=(0,1))
scaled = sc.fit_transform(data[sc_features])
data = pd.DataFrame(scaled,columns=sc.get_feature_names_out())
```

```python
data
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,dataset["SalePrice"],test_size=0.2,shuffle=True)
```

```python
gbc = GradientBoostingRegressor(max_depth=6)
```

```python
acc = []
for i in range(1,30):
    selector = SelectKBest(mutual_info_regression,k=i)
    selector.fit(x_train,y_train)
    sel_x_train = selector.transform(x_train)
    sel_x_test = selector.transform(x_test)
    gbc.fit(sel_x_train,y_train)
    acc.append(gbc.score(sel_x_test,y_test))
```

```python
acc
```

```python
selector = SelectKBest(mutual_info_regression,k=27)
selector.fit(x_train,y_train)
sel_x_train = selector.transform(x_train)
sel_x_test = selector.transform(x_test)
gbc.fit(sel_x_train,y_train)
print(gbc.score(sel_x_test,y_test))
```
