```python
# Making a copy of the original dataset to manipulate 
df = data.copy()
```

```python
df.dropna(how='any',inplace=True)
```

```python
df[df.duplicated()].shape[0]
```

```python
df = df.drop_duplicates()
```

```python
df.head(1)
```

```python
def convert_mileage(x):
    x = str(x)
    a = x.split(' ')
    if len(a) == 2:
        return float(a[0])
    try:
        return float(x)
    except:
        return None
```

```python
df['mileage'] = df['mileage'].apply(convert_mileage)
```

```python
def convert_power(x):
    x = str(x)
    a = x.split(' ')
    if len(a) == 2:
        return float(a[0])
    try:
        return x
    except:
        return None
```

```python
df['max_power'] = df['max_power'].apply(convert_power)
```

```python
df.drop('torque',axis=1,inplace=True)
```

```python
def convert_engine(x):
    x = str(x)
    a = x.split(' ')
    if len(a) == 2:
        return float(a[0])
    try:
        return float(x)
    except:
        return None
```

```python
df['engine'] = df['engine'].apply(convert_engine)
```
