```python
def cross_score(model,model_name):
    mean_score = cross_val_score(model,X=train_scaled.drop('remainder__Item_Outlet_Sales',axis=1),y=train_scaled['remainder__Item_Outlet_Sales'],cv=5).mean()
    print(f'Cross Val Score for {model_name} is: {round(mean_score,2)*100}%')

def train_model(model, model_name):
    model.fit(train_scaled.drop('remainder__Item_Outlet_Sales',axis=1),train_scaled['remainder__Item_Outlet_Sales'])
    model_train_score = model.score(train_scaled.drop('remainder__Item_Outlet_Sales',axis=1),train_scaled['remainder__Item_Outlet_Sales'])
    print(f"{model_name} model score on Training data: {round(model_train_score,2) * 100}%")
```
